```python
#стартовий код
import sys
from menu import Menu, MenuItem
from coffee_maker import CoffeeMaker
from money_machine import MoneyMachine

menu = Menu()
moneyMachine = MoneyMachine()
coffeeMaker = CoffeeMaker()

while True:
    user_input = input("Яку каву бажаєте? ({}): ".format(menu.get_items()))

    if user_input == "дякую":
        sys.exit()

    elif user_input == "чек":
        coffeeMaker.report()
        moneyMachine.report()

    elif menu.find_drink(user_input) != None:
        ordered = menu.find_drink(user_input)
        print("Ціна даної кави: {}".format(ordered.cost))
        if coffeeMaker.is_resource_sufficient(ordered):
            if moneyMachine.make_payment(ordered.cost):
                coffeeMaker.make_coffee(ordered)
```
