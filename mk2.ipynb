На внутрішньому ринку популярні піцерії пропонують до 15 **видів** піц, кожна із яких формується із
різних інгредієнтів, у залежності від рецепту - див. стовпчик МК-2 результатів захисту індивідуальних завдань і МК. 
Ваша піцерія організує автоматизацію обслуговування клієнтів, у залежності від власного
варіанту, який вказано у стовпчику МК-2 результатів захисту індивідуальних завдань і МК і відповідає ПІБ студента. Для кожного
студента із загального списку популярних піц, наданого за посиланням, для виконання МК-2 надано
лише 5 **видів** піц із відповідними номерами (назвами піц).
На основі принципів ООП, необхідно розробити і реалізувати застосунок, який забезпечує автоматизоване
обслуговування клієнтів, бажаючих замовити, в режимі самообслуговування, піцу із вказаного списку із 5 піц (згідно власного варіанту).
У процесі обслуговування клієнта необхідно використовувати наступні об'єкти (класи):
- МЕНЮ. Відповідальність об'єкту: взаємодія з клієнтом, вивід меню піцерії з
виводом назви піці і її номеру із загального списку, підтвердження замовлення піці із
вказаного меню, оплата з власного банківського рахунку. Для оплати замовлених піц (ЗАМОВЛЕННЯ) використовувати банківський акаунт, розроблений
в МК-1. Причому, наявними є три **типи** піці: стандарт, маленька і велика. Вага стандартної піці і її інгредієнтів задається виконавцем.
Причому, собівартість маленької піці 60% від стандартної піці з націнкою 5%, а собівартість великої піці 140% від стандартної піці з дисконтом 5% для кожного **виду** піці.
Собівартість кожного інгредієнту великої і маленької піці, розраховується на основі собівартості кожного інгредієнту піці стандарт, які задаються виконавцем.   
- ЗАМОВЛЕННЯ. Відповідальність об'єкту: містить список замовлених піц (не менше 3-х) у відповідності до собівартості
її інгредієнтів. 
- ПІЦА. Відповідальність об'єкту: базовий клас піц, відповідає заявленим узагальненим характеристикам
піц власного варіанту. 
- ПІЦА_назва+№. Відповідальність об'єкту: успадкований клас абстрактного базового класу ПІЦА, в якому здійснюється уточнення 
характеристик (атрибутів, інгредієнтів і методів) кожного **виду** піці. Ім'я кожного успадкованого класу ПІЦА_назва+№ відповідає назві і номеру піці, згідно власного варіанту.
Таких успадкованих класів піц - 5, згідно номерів **видів** піц власного варіанту. 

## Завдання:


1. Автоматизований термінал обслуговування клієнта повинен відображати `МЕНЮ` піц із 5 піц власного варіанту.  
2. МЕНЮ дозволяє в автоматизованому варіанті створити `ЗАМОВЛЕННЯ`, з урахуванням **типу** і **виду** піці. 
3. Клієнт формує `ЗАМОВЛЕННЯ` шляхом введення номера піці із `МЕНЮ`.
4. `ЗАМОВЛЕННЯ` формується у вигляді списку із номерів піц, представлених в `МЕНЮ`.
5. У `ЗАМОВЛЕННІ` повинно бути не менше 3-х **видів** піц обраного типу.
6. Клієнт підтверджує `ЗАМОВЛЕННЯ` обраних піц або може відмінити `ЗАМОВЛЕННЯ`.
7. Клієнт отримує рахунок за `ЗАМОВЛЕННЯ` і може перевірити `СТАТУС` банківського рахунку до оплати `ЗАМОВЛЕННЯ` із автоматизованим визначенням можливості оплати цього `ЗАМОВЛЕННЯ`.
8. Клієнт сплачує `ЗАМОВЛЕННЯ` з власного банківського рахунку.
9.  Клієнт отримує чек про оплату `ЗАМОВЛЕННЯ`: **типи** і **види** замовлених піц, **ціна** кожної і загальна **сума**.
10. Клієнт отримує статус власного банківського рахунку після оплати `ЗАМОВЛЕННЯ`.

```python
### Pizza###

from abc import ABC, abstractmethod

class Pizza(ABC):
    @abstractmethod
    def id(self):
        pass

    @abstractmethod
    def name(self):
        pass

    @abstractmethod
    def ingredients(self):
        pass
    @abstractmethod
    def price(self):
        pass

    @abstractmethod
    def weight(self):
        pass

class Pizza_Hawaiian4(Pizza):
    def __init__(self):
        self._id = 4
        self._name = "Гавайська піца"
        self._weight = 600
        self._price = 169
        self._ingredients = ["шинка", "ананас", "цибуля", "зелений перець", "халопеньйо", "сир", "томатний соус"]

    @property
    def id(self):
        return self._id

    @property
    def name(self):
        return self._name

    @property
    def weight(self):
        return self._weight

    @property
    def price(self):
        return self._price

    @property
    def ingredients(self):
        return self._ingredients

class Pizza_Calzone6(Pizza):
    def __init__(self):
        self._id = 6
        self._name = "Піца Кальцоне"
        self._weight = 430
        self._price = 145
        self._ingredients = ["орегано", "шинка", "гриби", "рикотта", "моцарела", "пармезан"]

    @property
    def id(self):
        return self._id

    @property
    def name(self):
        return self._name

    @property
    def weight(self):
        return self._weight

    @property
    def price(self):
        return self._price

    @property
    def ingredients(self):
        return self._ingredients


class Pizza_With_Seafood8(Pizza):
    def __init__(self):
        self._id = 8
        self._name = "Піца з морепродуктами"
        self._weight = 600
        self._price = 229
        self._ingredients = ["креветки", "мідії", "часниковий соус", "лимон", "оливки", "маслини"]

    @property
    def id(self):
        return self._id

    @property
    def name(self):
        return self._name

    @property
    def weight(self):
        return self._weight

    @property
    def price(self):
        return self._price

    @property
    def ingredients(self):
        return self._ingredients

class Pizza_Vegetable11(Pizza):
    def __init__(self):
        self._id = 11
        self._name = "Овочева піца"
        self._weight = 600
        self._price = 129
        self._ingredients = ["гриби", "солодкий перець", "кримська цибуля", "маслини", "томати", "томатний соус"]

    @property
    def id(self):
        return self._id

    @property
    def name(self):
        return self._name

    @property
    def weight(self):
        return self._weight

    @property
    def price(self):
        return self._price

    @property
    def ingredients(self):
        return self._ingredients


class Pizza_With_Tuna12(Pizza):
    def __init__(self):
        self._id = 12
        self._name = "Піца з тунцем"
        self._weight = 600
        self._price = 209
        self._ingredients = ["томатний соус", "маслини", "моцарела", "тунець"]

    @property
    def id(self):
        return self._id

    @property
    def name(self):
        return self._name

    @property
    def weight(self):
        return self._weight

    @property
    def price(self):
        return self._price

    @property
    def ingredients(self):
        return self._ingredients
```

```python
### Bank Account ###

class BankAccount:
    def __init__(self, first_name, second_name, status=0):
        self.first_name = first_name
        self.second_name = second_name
        self.status = status

    def display_account(self):
        print("{} {} {}".format(self.first_name, self.second_name, self.status))

    def put_account(self, amount_of_money):
        self.status += amount_of_money

    def get_status(self):
        return self.status

    def get_account(self, amount_of_money=None):
        if amount_of_money == None:
            amount_of_money = self.status
        if amount_of_money > self.status:
            raise ValueError("Not enough money on account")
        self.status -= amount_of_money
        return amount_of_money

    def send_deposit(self, amount_of_money, deposit_rate, deposit_time):
        deposit_tax = 19.5
        return round(amount_of_money + (amount_of_money*(1 + deposit_rate/100)**deposit_time - amount_of_money)*(1 - deposit_tax/100),2)
```

```python
### Order ###

class Order:
    def __init__(self):
        self.order_content = list()

    def add_to_order(self, order):
        self.order_content.append(order)

    def __len__(self):
        return len(self.order_content)
    def __str__(self):
        s = ""
        for i in self.order_content:
            if i[1] == 1:
                s += "{} {}\n".format(i[0].name, "стандарт")
            elif i[1] == 2:
                s += "{} {}\n".format(i[0].name, "мала")
            elif i[1] == 3:
                s += "{} {}\n".format(i[0].name, "велика")
        return s

    def get_total_price(self):
        total_price = 0
        for order in self.order_content:
            if order[1] == 1:
                total_price += round(order[0].price, 2)
            elif order[1] == 2:
                total_price += round(order[0].price * 1.05 * 0.6, 2)

            elif order[1] == 3:
                total_price += round(order[0].price * 0.95 * 1.4, 2)

        return total_price


    def get_check(self):
        price = 0
        t = ""
        print("\nВаш чек")
        for order in self.order_content:
            if order[1] == 1:
                price =  round(order[0].price, 2)
                t = "стандарт"
            elif order[1] == 2:
                price =  round(order[0].price * 1.05 * 0.6, 2)
                t = "міні"

            elif order[1] == 3:
                price =  round(order[0].price * 0.95 * 1.4, 2)
                t = "велика"
            print("{} {}\n\tЦіна: {}".format(t, order[0].name, price))

        print("Загальна ціна:", self.get_total_price())
```
