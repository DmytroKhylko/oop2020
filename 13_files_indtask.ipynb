```python
# стартовий код
from abc import ABC, abstractmethod
import math
# для застосування стандартної функції sqrt 
class Quarter:
    @abstractmethod
    def quarter(self):
        pass

class Point(Quarter):
    """ клас об'єктів, що визначають крапки на площині
    """
    
    def __init__(self, init_x, init_y):
        """ конструктор для класу об'єктів, який визначає крапку 
            з координатами init_x, init_y)
        """
        self._x = init_x
        self._y = init_y
        
    @property
    def x(self):
        return self._x

    @property
    def y(self):
        return self._y

    @x.setter
    def x(self, x):
        self._x = x


    @y.setter
    def y(self, y):
        self._y = y

    def distance_from_start(self):
        """ визначає дистанцію крапки 
            (self) від початку координат(0, 0))
        """
        return math.sqrt((self._x - 0)**2 + (self._y - 0)**2)

    def move_down(self, amount):
        """ зсуває крапку (self) вниз 
            на визначену величину 
        """
        self._y -= amount

    def quarter(self):
        if self._x == 0 or self._y == 0:
            return "NO"
        if self._x > 0:
            if self._y > 0:
                return "I"
            return "IV"
        if self._y > 0:
            return "II"
        return "III"

    def change(self):
        self._x = -1 * self._x
        self._y = -1 * self._y

    def __str__(self):
        return "({}, {})".format(self._x, self._y)

    def __eq__(self, other):
        if self._x == other._x and self._y == other._y:
            return True
        return False

    def this_quarter(self, other):
        if self.quarter() == other.quarter():
            return True
        return False



krapka1 = Point(6,0)

krapka2 = Point(0,-8)
print(krapka1)
print(krapka2)

print("Відстань від крапки {} до початку координат дорівнює {}".format(krapka1, krapka1.distance_from_start()))
print("Відстань від крапки {} до початку координат дорівнює {}".format(krapka2, krapka2.distance_from_start()))

krapka1.move_down(5)
print("krapka1 була зміщена на 5 одиниць вниз", krapka1)

print(krapka1.quarter())


print(krapka1 == krapka2)

print(krapka2)
krapka2.change()
print(krapka2)

print(krapka1.this_quarter(krapka2))
```
